```python
import feedparser
import requests
```

```python
feed = feedparser.parse("https://www.coindesk.com/feed/")
feed
```

```python
feed['entries']
```

```python
len(feed['entries'])
```

```python
feed['entries'][0]
```

```python
coindesk_urls = []

for entry in feed['entries']:
    coindesk_urls.append(entry['link'])
```

```python
coindesk_urls
```

```python
for url in coindesk_urls:
    with open("coindesk._news.txt", "a", encoding="utf8") as f:
        f.write(url+"\n")
```

```python
from bs4 import BeautifulSoup as bs
req = requests.get(coindesk_urls[0])
soup = bs(req.text, 'html.parser')
soup_body = soup.find("section", class_ = "has-media news article-body")
```

```python
soup_body.find("em")
```

```python
from lxml import html
```

```python
lxml_html = html.fromstring(req.text)
lxml_html
```

```python
lxml_html.find_class("article-content-container noskimwords")
```

```python
lxml_end_point = lxml_html.cssselect("em")
lxml_end_point
```

```python
lxml_end_point[0]
```

```python
end_parent = lxml_end_point.getparent() 
```

```python
body_lxml = []

for elem in end_parent.itersiblings(preceding=True):
    body_lxml.append(elem)
```
